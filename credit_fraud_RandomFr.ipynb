## Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
#load the dataset
df = pd.read_csv('data.csv')
df.head()
```

```python
df.columns
```

```python
# from the get go we know that the Application Id is useless when it comes to predicting the outcome so it is safe to say that we can drop the same.
df.drop('Application ID',inplace=True,axis=1)
```

```python
#show the first 5 entries
df.head()
```

```python
# Lets see how our target variable has been distributed.
sns.countplot(df['Performance Tag'])

#this is a major issue as it will lead to biased models. We will see in the calculations
```

```python
df['Performance Tag'].value_counts()
```

```python
# we have a lot of categorical data,lets convert that into numerical format using one hot encoding.
df_num = pd.get_dummies(data=df, columns=['Gender', 'Marital Status (at the time of application)','Education','Profession ','Type of residence'],drop_first=True)
```

```python
df_num.head()
```

```python
df_num.columns
```

```python
df_num.isnull().sum()
```

```python
#dropping the missing rows as they are less than 5%
df_num.dropna(inplace=True)
```

```python
#let me define the target and the independent variables by storing them in different dataframes.

X = df_num.drop('Performance Tag',axis=1)
y = df_num['Performance Tag']
```

## Using the train_test_split

```python
#lets distribute our data into train and test split:
from sklearn.model_selection import train_test_split
```

```python
#the test set has a size of 20% of the dataset if we set the test_size=0.2, and setting the random_state ensures that we get the same split:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=102)
```

```python
from sklearn.ensemble import RandomForestClassifier
rcf = RandomForestClassifier()
```

```python
rcf.fit(X_train,y_train)
```

```python
#checking the predictions done by the model.
predicted = rcf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,predicted))
print('\n')
print(classification_report(y_test,predicted))
```

## The model has not succeded in predicting the people who have defaulted (Performance tag = 1 ). So ensemble techniques are always not effective if done on imbalanced sets.

The value of 0 corresponds to non defaulters and 1 refers to the ones who have defaulted.

## lets use undersampling to level our datset from imblearn library:
note one thing that this is just oversimplification of our problem. i just wanted too showcase how imbalanced sets can pose an issue.

```python
from imblearn.under_sampling import NearMiss
nm = NearMiss()
X_res,y_res = nm.fit_sample(X,y)
```

```python
X_res.shape
```

```python
y_res.shape
```

```python
#the test set has a size of 20% of the dataset if we set the test_size=0.2, and setting the random_state ensures that we get the same split:
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.20, random_state=102)
```

```python
rcf.fit(X_train,y_train)
```

```python
#checking the predictions done by the model.
predicted = rcf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(y_test,predicted))
print('\n')
print(classification_report(y_test,predicted))
```

### This is a great accuracy score (or maybe not) depending on how much accuracy you were expecting. Also, note that the type-2 error is less here as compared to type-1.
